```python
import requests

from bs4 import BeautifulSoup
import pandas as pd
import time
```

# Scraping Data

```python
url = "https://en.wikipedia.org/wiki/List_of_best-selling_albums"
page = requests.get(url)
soup = BeautifulSoup(page.text)
```

```python
soup.find('table')
```

```python
soup.find_all('table')
```

```python
soup.find('table', class_ = 'wikitable')
```

```python
table = soup.find_all('table')[1]
print(table)
```

```python
titles = table.find_all('th')
print(titles)
```

```python
album_titles = [title.text.strip() for title in titles]
print(album_titles)
```

```python
df = pd.DataFrame(columns = album_titles)
print(df)
```

```python
column_data = table.find_all("tr")
```

```python
for row in column_data[1:] :
    row_data = row.find_all('td')
    indiv_row_data = [data.text.strip() for data in row_data]

    length = len(df)
    df.loc[length] = indiv_row_data
```

```python
print(df)
```

# Prepocessing dan Hasil

```python
data = pd.DataFrame(df)
data.head(5)
```

Tabel di atas merupakan hasil scraping yang diperoleh. Namun, total certified copies yang ingin kami lihat hanyalah total secara keseluruhan. Selain itu, kami tidak membutuhkan informasi referensi. Dengan demikian, akan dilakukan prepocessing dengan tujuan tersebut.

```python
data['Total certified copies(from available markets)*'] = data['Total certified copies(from available markets)*'].str.split('\n').str[0]
data.drop('Ref(s)', axis=1, inplace=True)
```
