# I. Multidimensional Scaling

## B. Metric Muldimensional Scaling

![](https://i.postimg.cc/13Qj4Myj/data1.png)

### 1. Classical multidimensional scaling (MDS) of a data matrix. Also known as principal coordinates analysis (Gower, 1966).

````r
# Cmpute MDS
data11 <- as.dist(data1) #ubah jd matriks distance buku pers 16.1
mds11 <- data11 %>% cmdscale()
colnames(mds11) <- c("Dim.1", "Dim.2")
mds11
````

````r
options(repr.plot.width = 10, repr.plot.height = 7, repr.plot.res = 200)
# Plot MDS
ggscatter(as_tibble(mds11), x = "Dim.1", y = "Dim.2",
          label = rownames(data1),
          size = 1,
          repel = TRUE) + geom_vline(xintercept = 0) + geom_hline(yintercept=0)
````

### 2. Multidimensional scaling on a symmetric dissimilarity matrix using SMACOF.

````r
mds12 <- mds(data11,ndim = 2, type = "ratio" )
mds12
````

````r
summary(mds12) #0-5%: Sangat baik, menunjukkan representasi yang sangat akurat.stress per point
````

````r
plot(mds12, type = "p", label.conf = list(label = TRUE, col = "darkgray"), pch = 25, col = "red")
````

````r
scree.plot = function(d, k) {
    stresses=sammon(d, k=k)$stress
    for(i in rev(seq(k-1)))
        stresses=append(stresses,sammon(d, k=i)$stress)
    plot(seq(k),rev(stresses), type="b", xaxp=c(1,k, k-1), ylab="Stress", xlab="Number of dimensions")
    }
scree.plot(data22, k=5)
````

## B. Multiple Correspondence Analysis

![](https://i.postimg.cc/HW05dtPx/data4.png)

### 1. Preprocessing Data

````r
head(data4)
dim(data4)
````

````r
df <- data.frame(Dust=factor(),
                 Race=factor(),
                 Sex=factor(),
                 Smoking=factor(),
                 Work=factor(),
                 Suffer_Byssinosis=factor(),
                 stringsAsFactors=FALSE)

gab_yes <- mapply(list, 1:dim(data4)[1], mapply(list, data4[1], data4[2], SIMPLIFY=F)$Yes[[1]], SIMPLIFY=F)
gab_no <- mapply(list, 1:dim(data4)[1], mapply(list, data4[1], data4[2], SIMPLIFY=F)$Yes[[2]], SIMPLIFY=F)

data4_final = data4[3:7]

for (i in 1:dim(data4)[1]) {
    if ((gab_yes[[i]][2]) == 0) {next}
    a <- data4_final[rep(as.integer(gab_yes[[i]][1]), as.integer(gab_yes[[i]][2])),]
    a$Suffer_Byssinosis <- as.factor('Yes')
    df <- rbind(df, a)
}

for (i in 1:dim(data4)[1]) {
    if ((gab_no[[i]][2]) == 0) {next}
    a <- data4_final[rep(as.integer(gab_no[[i]][1]), as.integer(gab_no[[i]][2])),]
    a$Suffer_Byssinosis <- as.factor('No')
    df <- rbind(df, a)
}

df
````

````r
dim(df)
df <- as.data.frame(df)
rownames(df) <- 1:dim(df)[1]
````

### 2. Melakukan MCA

````r
str(df)
df$Dust <- as.factor(df$Dust)
df$Race <- as.factor(df$Race)
df$Sex <- as.factor(df$Sex)
df$Smoking <- as.factor(df$Smoking)
df$Work <- as.factor(df$Work)
str(df)
````

````r
mca <- MCA(df, graph=TRUE)
mca
````

````r
# table of eigenvalues
mca$eig
````

````r
# column coordinates
head(mca$var$coord)
````

````r
# row coordinates
head(mca$ind$coord)
````

````r
# number of categories per variable
cats = apply(df, 2, function(x) nlevels(as.factor(x)))
cats
````

````r
# data frames for ggplot
mca_vars_df = data.frame(mca$var$coord, Variable = rep(names(cats),
    cats))
mca_obs_df = data.frame(mca$ind$coord)

# plot of variable categories
ggplot(data = mca_vars_df, aes(x = Dim.1, y = Dim.2, label = rownames(mca_vars_df))) +
    geom_hline(yintercept = 0, colour = "gray70") + geom_vline(xintercept = 0,
    colour = "gray70") + geom_text(aes(colour = Variable)) + ggtitle("MCA plot of variables using R package FactoMineR")
````

````r
# MCA plot of observations and categories
ggplot(data = mca_obs_df, aes(x = Dim.1, y = Dim.2)) + geom_hline(yintercept = 0,
    colour = "gray70") + geom_vline(xintercept = 0, colour = "gray70") + geom_point(colour = "gray50",
    alpha = 0.7) + geom_density2d(colour = "gray80") + geom_text(data = mca_vars_df,
    aes(x = Dim.1, y = Dim.2, label = rownames(mca_vars_df), colour = Variable)) +
    ggtitle("MCA plot of variables using R package FactoMineR") + scale_colour_discrete(name = "Variable")
````
