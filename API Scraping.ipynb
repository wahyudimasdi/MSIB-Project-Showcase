## Harga Emas

```python
# Scraping Harga Beli Emas 7 Tahun Terakhir pada Website www.indogold.id

url = "https://www.indogold.id/ajax/chart_range/GOLD/2017-03-31/2024-03-31"

payload = {}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)
print(response.text)
```

```python
# Convert ke json
beliemas_js = json.loads(response.text)
beliemas_js
```

```python
# Normalisasi data
beliemas_normal = pd.json_normalize(beliemas_js)
beliemas_normal
```

```python
# Membuat dataframe harga emas

# Memasukkan data harga beli emas ke dataframe dengan nama kolom 'Harga_Beli_Emas'
hargaemas = pd.DataFrame(beliemas_normal['data'].tolist()).transpose()
hargaemas.columns = ['Harga_Beli_Emas']
hargaemas['Harga_Beli_Emas'] = hargaemas['Harga_Beli_Emas'].apply(lambda x: x[1])

# Memberi tanggal dari 31 Maret 2017 hingga 31 Maret 2024
tanggal = pd.DataFrame({'Tanggal': pd.date_range(end=datetime(2024,3,31), periods=len(hargaemas), freq='D')})
hargaemas = pd.concat([tanggal, hargaemas], axis=1)
hargaemas
```

```python
# Scraping Harga Jual Emas 7 Tahun Terakhir pada Website www.indogold.id

url = "https://www.indogold.id/ajax/chart_range_jual/GOLD/2017-03-31/2024-03-31"

payload = {}
headers = {}

response = requests.request("GET", url, headers=headers, data=payload)
print(response.text)
```

```python
# Convert ke json
jualemas_js= json.loads(response.text)
jualemas_js
```

```python
# Normalisasi data
jualemas_norm = pd.json_normalize(jualemas_js)
jualemas_norm
```

```python
# Memasukkan data harga jual emas ke dataframe hargaemas
hargaemas['Harga_Jual_Emas'] = pd.DataFrame(jualemas_norm['data'].tolist()).transpose()
hargaemas['Harga_Jual_Emas'] = hargaemas['Harga_Jual_Emas'].apply(lambda x: x[1])

# Didapatkan Tabel Harga Beli dan Harga Jual Emas per Gram dalam 7 Tahun Terakhir
hargaemas
```

# FLIGHTSTATS -> Info Detail Penerbangan

```python
import requests
url = "https://www.flightstats.com/v2/api-next/flight-tracker/arr/CGK/2024/3/29/6?carrierCode=&numHours=6&rqid=xw84eh8f3a"

payload={}
headers = {"User-Agent" : "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"}

response = requests.request("GET", url, headers=headers, data=payload)
print(response.text)
```

```python
# Ganti tipe ke JSON
import json

json_data = json.loads(response.text)['data']['flights']
json_data
```

```python
# Tabulasi
import pandas as pd

df = pd.json_normalize(json_data)
df
```

```python
list_bandara = ['CGK', 'BDO', 'DPS', 'SUB', 'UPG']

list_df = []

# Iterasi bandara
for bandara in list_bandara:
    # Requests API
    url = f"https://www.flightstats.com/v2/api-next/flight-tracker/arr/{bandara}/2024/3/29/6?carrierCode=&numHours=6&rqid=xw84eh8f3a"
    response = requests.request("GET", url, headers=headers, data=payload)

    # Ubah JSON
    json_data = json.loads(response.text)['data']['flights']

    # Ubah menjadi DF
    df = pd.json_normalize(json_data)

    print(f'Penerbangan dari {bandara} sebanyak {len(df)}.')

    # Input df hasil request ke list
    list_df.append(df)

df_hasil = pd.concat(list_df, ignore_index=True)
df_hasil
```

Code tersebut digunakan untuk melacak penerbangan kedatangan di bandara tertentu pada tanggal 29 Maret 2024, pukul 6 pagi waktu lokal, dengan informasi yang mencakup nomor penerbangan, maskapai, waktu kedatangan, asal penerbangan, dan status penerbangan.Data yang diambil adalah data dari 5 bandara di indonesia, yaitu:
•	CGK: Bandara Internasional Soekarno-Hatta (Jakarta, Indonesia)
•	BDO: Bandara Internasional Husein Sastranegara (Bandung, Indonesia)
•	DPS: Bandara Internasional Ngurah Rai (Denpasar, Bali, Indonesia)
•	SUB: Bandara Internasional Juanda (Surabaya, Indonesia)
•	UPG: Bandara Internasional Sultan Hasanuddin (Makassar, Indonesia)
Berikut adalah penjelasan untuk masing-masing variabel yang terdapat dalam dataset:
1.	`sortTime`: Waktu pengurutan, menunjukkan waktu ketika data tersebut diurutkan dalam dataset. Biasanya digunakan untuk tujuan pengurutan dan tidak memiliki makna intrinsik terkait data penerbangan itu sendiri.
2.	`operatedBy`: Operator penerbangan atau maskapai yang mengoperasikan penerbangan.
3.	`url`: URL yang mengarah ke informasi lebih lanjut tentang penerbangan. Ini dapat digunakan untuk mengakses detail seperti status penerbangan, waktu tiba, dan lain-lain.
4.	`departureTime.timeAMPM`: Waktu keberangkatan dalam format AM/PM (Ante Meridiem/Post Meridiem).
5.	`departureTime.time24`: Waktu keberangkatan dalam format 24 jam (format militer).
6.	`arrivalTime.timeAMPM`: Waktu kedatangan dalam format AM/PM (Ante Meridiem/Post Meridiem).
7.	`arrivalTime.time24`: Waktu kedatangan dalam format 24 jam (format militer).
8.	`carrier.fs`: Kode IATA dari maskapai penerbangan.
9.	`carrier.name`: Nama maskapai penerbangan.
10.	`carrier.flightNumber`: Nomor penerbangan yang dioperasikan oleh maskapai tersebut.
11.	`airport.fs`: Kode IATA dari bandara tujuan.
12.	`airport.city`: Kota tempat bandara tujuan berlokasi.
13.	`isCodeshare`: Indikator apakah penerbangan ini adalah bagian dari kesepakatan codeshare antara maskapai. Jika iya, akan memiliki nilai, jika tidak, mungkin bernilai NaN atau False.



# FLIGHTRADAR24 -> Info Daftar Penerbangan

```python
import requests

url = "https://www.flightradar24.com/webapi/v1/airport-disruptions?continent=0&period=live&type=both&indices=false"
params= {}
headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/123.0.0.0 Safari/537.36"}

response = requests.request("GET", url, params=params, headers=headers)

print(response.text)
```

```python
import pandas as pd
import json
json_data = json.loads(response.text)['data']['rank']
json_data
```

```python
df = pd.json_normalize(json_data)
df
```

```python
flattened_data_full = []

for record in json_data:
    airport_data_full = {
        'iata': record['airport']['code']['iata'],
        'icao': record['airport']['code']['icao'],
        'airport_name': record['airport']['name'],
        'city': record['airport']['city'],
        'country_name': record['airport']['country']['name'],
        'arrivals_today_total': record['arrivals']['today']['total'],
        'departures_today_total': record['departures']['today']['total']
    }
    flattened_data_full.append(airport_data_full)

# Convert the flattened data into a DataFrame
df_complete_full = pd.DataFrame(flattened_data_full)

df_complete_full
```

Code tersebut digunakan untuk mengambil data tentang gangguan di bandara secara langsung dari Flightradar24, dengan parameter yang mencakup periode waktu "live", tipe gangguan baik itu kedatangan maupun keberangkatan, dan pengaturan lainnya seperti pembagi indeks yang diaktifkan.
Berikut adalah penjelasan untuk masing-masing variabel yang terdapat dalam dataset:
1.	iata: Kode IATA dari bandara.
2.	icao: Kode ICAO dari bandara.
3.	airport_name: Nama lengkap dari bandara.
4.	city: Kota dimana bandara tersebut terletak.
5.	country_name: Nama negara dimana bandara tersebut terletak.
6.	arrivals_today_total: Total kedatangan penerbangan pada hari tersebut di bandara tersebut.
7.	departures_today_total: Total keberangkatan penerbangan pada hari tersebut dari bandara tersebut.
