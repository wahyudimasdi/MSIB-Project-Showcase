# Insight 1

Akan dilihat kecenderungan atau trend angkatan lama vs angkatan baru yang memilih untuk membuka usaha sendiri (wirausaha) dibanding dengan bekerja bada suatu perusahaan. Ingin dicari tau kenapa angkatan tersebut cenderung memilih wirausaha dibanding bekerja pada suatu perusahaan.

```python
data['Tahun Wisuda'] = data['Tahun Wisuda'].astype(int)
```

```python
dw = data['Tahun saudara mulai berwirausaha:'].fillna(0)
dw = dw.where(dw == 0, 1)
data1 = {"lulusan" : data['Tahun Wisuda'],
         "Wirausaha" : dw,
         "gaji" : data['Perkiraan rata-rata penghasilan (take-home-pay) anda per-bulan saat ini (dalam juta rupiah):']}
data1 = pd.DataFrame(data1)
data1
```

```python
data1 = data1.dropna(subset=['gaji'])
data1
```

```python
import matplotlib.pyplot as plt

# Memisahkan data angkatan baru (tahun 2020 ke atas) dan angkatan lama (tahun sebelumnya)
angkatan_baru = data1[data1['lulusan'] >= 2020]
angkatan_lama = data1[data1['lulusan'] < 2020]

# Menghitung jumlah mahasiswa yang memilih wirausaha per tahun angkatan untuk kedua kelompok
wirausaha_count_baru = angkatan_baru.groupby('lulusan')['Wirausaha'].sum().reset_index()
wirausaha_count_lama = angkatan_lama.groupby('lulusan')['Wirausaha'].sum().reset_index()

# Membuat dua line chart terpisah
plt.figure(figsize=(12, 5))

#Lulusan lama
plt.subplot(1, 2, 2)
plt.plot(wirausaha_count_lama['lulusan'], wirausaha_count_lama['Wirausaha'], marker='o', linestyle='-')
plt.xlabel('Lulusan Lama')
plt.ylabel('Jumlah Mahasiswa Memilih Wirausaha')
plt.title('Tren Memilih Wirausaha per Tahun Angkatan (Lulusan Lama)')
plt.grid(True)
plt.xticks(range(min(wirausaha_count_lama['lulusan']), max(wirausaha_count_lama['lulusan'])+1, 1),  rotation=45)

#Lulusan Baru
plt.subplot(1, 2, 1)
plt.plot(wirausaha_count_baru['lulusan'], wirausaha_count_baru['Wirausaha'], marker='o', linestyle='-')
plt.xlabel('Lulusan Baru')
plt.ylabel('Jumlah Mahasiswa Memilih Wirausaha')
plt.title('Tren Memilih Wirausaha per Tahun Angkatan (Lulusan Baru)')
plt.grid(True)
plt.xticks(range(min(wirausaha_count_baru['lulusan']), max(wirausaha_count_baru['lulusan'])+1, 1),  rotation=45)

plt.tight_layout()
plt.show()
```

# Insight 2

Mau lihat perbandingan gaji saat ini untuk setiap lulusan S1 pada setiap tahunnya. Apakah benar gaji fresh graduate tanpa harus lanjut studi bisa menyaingi gaji lulusan tahun lama karena perkembangan IPTEK.

```python
df = data['Nama jurusan/program studi setelah S1'].fillna(0)
df = df.where(df == 0, 1)
data2 = {"waktu lulus S1" : data['Tahun Wisuda'],
         "apakah lanjut S2" : df,
         "Waktu Mendapat Pekerjaan lulusan s1" : data["Waktu Mendapatkan Pekerjaan:"],
         "gaji" : data['Perkiraan rata-rata penghasilan (take-home-pay) anda per-bulan saat ini (dalam juta rupiah):']}
data2 = pd.DataFrame(data2)
data2
```

```python
data2 = data2[~(data2['apakah lanjut S2'] == 1)]
data2
```

```python
data2 = data2.dropna(subset=['gaji'])
data2
```

```python
plt.figure(figsize=(10, 6))
plt.bar(data2["waktu lulus S1"], data2["gaji"], color="skyblue", ec="black")

plt.xlabel('Tahun Lulus')
plt.ylabel('Gaji (jt)')
plt.title('Gaji Lulusan Tahun X')

plt.grid(True)
plt.show()
```

```python
# Mengelompokkan data berdasarkan tahun dan menghitung rata-rata gaji
df_avg = data2.groupby('waktu lulus S1')['gaji'].mean().reset_index()

# Plot bar chart dengan rata-rata gaji per tahun
plt.figure(figsize=(10, 6))
plt.bar(df_avg['waktu lulus S1'], df_avg['gaji'], color='orange', ec="black")


plt.xlabel('Tahun Lulus')
plt.ylabel('Rata-rata Gaji (jt)')
plt.title('Rata-rata Gaji Lulusan Tahun X')

plt.grid(True)
plt.show()
```

# Insight 3

Perbandingan penghasilan lulusan s1, s2/s3, dan wirausahawan. Lalu kelompok gaji yang tertinggi ingin dilihat apasih posisinya dalam pekerjaan saat ini

```python
d23 = data['Nama jurusan/program studi setelah S1'].fillna(0)
d23 = d23.where(d23 == 0, 1)
dw = data['Tahun saudara mulai berwirausaha:'].fillna(0)
dw = dw.where(dw == 0, 1)

data3 = { "Lanjut studi" : d23,
         "Wirausaha" : dw,
          "posisi" : data["Posisi anda saat ini di tempat bekerja saat Ini:"],
         "gaji" : data['Perkiraan rata-rata penghasilan (take-home-pay) anda per-bulan saat ini (dalam juta rupiah):']}

data3 = pd.DataFrame(data3)
data3
```

```python
data3 = data3.dropna(subset=['gaji'])
data3
```

```python
lanjut_studi = data3[data3['Lanjut studi'] == 1]
tidak_lanjut_studi = data3[data3['Lanjut studi'] == 0]
wirausaha = data3[data3['Wirausaha'] == 1.0]

# Menghitung rata-rata gaji untuk masing-masing kelompok
avg_salary_lanjut_studi = lanjut_studi['gaji'].mean()
avg_salary_tidak_lanjut_studi = tidak_lanjut_studi['gaji'].mean()
avg_salary_wirausaha = wirausaha['gaji'].mean()

# Menyiapkan data untuk visualisasi
categories = ['Lanjut Studi', 'Tidak Lanjut Studi', 'Wirausaha']
average_salaries = [avg_salary_lanjut_studi, avg_salary_tidak_lanjut_studi, avg_salary_wirausaha]

# grafik
plt.figure(figsize=(8, 5))
plt.bar(categories, average_salaries, color=['blue', 'green', 'red'])
plt.xlabel('Kelompok Mahasiswa')
plt.ylabel('Gaji Rata-rata')
plt.title('Perbandingan Gaji Rata-rata Mahasiswa')
plt.show()
```

```python
# Lalu kelompok gaji yang tertinggi ingin dilihat apasih posisinya dalam pekerjaan saat ini
data3 = data3.drop(columns=['Wirausaha'])
data3 = data3[data3['Lanjut studi'] == 1]
data3
```

```python
import plotly.express as px
# Menghitung jumlah value untuk setiap kategori dalam variabel "Posisi"
position_counts = data3['posisi'].value_counts().reset_index()
position_counts.columns = ['posisi', 'Jumlah']

fig = px.pie(position_counts, values='Jumlah', names='posisi', title='Posisi Pekerjaan Lulusan S2 dan S3', color_discrete_sequence=px.colors.sequential.RdBu) #ganti sama yang varnya gede nanti
fig.update_layout(
    title_text='Pie Chart Posisi Pekerjaan',
    title_x=0.5,  # Menyusun judul horizontal di tengah
    title_y=0.95  # Menyusun judul secara vertikal di atas pie chart
    )
fig.show()
```
